```python
json_dict = {}
no_data = []
for tick_bbg, row in tickers_dict.loc[tickers_todownload['0']].iterrows():
    tick = row['ticker']
    if not type(tick) is str:
        continue
#         print(tick)
#         i = input('Please input num')
#         tick = row['ticker'].iloc[i]
    df = DL.loadDaily(tick)
    if len(df) == 0:
        no_data.append(tick_bbg)
        continue
    date_index = list(df.index)
    date_index = [datetime.strftime(x, '%Y-%m-%d') for x in date_index]
    json_dict[tick_bbg] = date_index
```
